# Project:Exploratory Data Analysis on Drugs, Side Effects, and Medical Conditions

```python
import pandas as pd
```

```python
raw_data = pd.read_csv("drugs_side_effects_drugs_com.csv")
df = raw_data.copy()
df.head()
```

### Handling Missing Values in the `rating` Column by Replacing with Mean

```python
df['rating'] = df['rating'].fillna(df['rating'].mean())
```

```python
df['rating']
```

#### Handling Missing Values in the `no_of_rating` Column by Replacing with Min

```python
df['no_of_reviews'] = df['no_of_reviews'].fillna(df['no_of_reviews'].min())
```

#### Replacing Missing Values with 'Unknown' in the DataFrame

```python
df.fillna('unknown',inplace = True)
```

```python
df.isnull().sum()
```

### Chaning Datatype of `Activity` Column

```python
df['activity'].dtype
```

```python
df['activity'] = df['activity'].replace('%', '', regex=True).astype(int)
```

```python
df['activity'].dtype
```

### create a checkpoint

```python
df_cleaned = df.copy()
```

## EDA

#### Count the most common drugs for each medical condition

```python
df_cleaned.head(2)
```

### Top 10 Most Frequently Prescribed Drugs Across Medical Conditions

```python
top_drugs =df.groupby('medical_condition')['drug_name'].value_counts().nlargest(10)
top_drugs
```

```python
top_drugs.plot(kind = 'bar')
```

### Count of Drugs by Medical Condition by each medical condition

```python
count_of_drugs = df_cleaned.groupby('medical_condition')['drug_name'].count().sort_values(ascending=True )
count_of_drugs
```

```python
count_of_drugs.plot(kind = 'bar',figsize=(12,6),color = 'green')
```

```python
top_50_drug_class = df_cleaned['drug_classes'].value_counts().sort_values(ascending = False).head(50)
top_50_drug_class
```

```python
top_50_drug_class.plot(kind = 'barh',figsize = (12,12))
```
