## Working with Reasons for Absence column
Split the reasons for absence into multiple dummy variables, and then group them in the following way:

Group 1: Columns 1 to 14

Group 2: Columns 15, 16, and 17

Group 3: Columns 18, 19, 20, and 21

Group 4: Columns 22 to 28

```python
df["Reason for Absence"].min()
```

```python
df["Reason for Absence"].max()
```

```python
sorted(df["Reason for Absence"].unique())
```

```python
#checking the length of the unique values
len(df["Reason for Absence"].unique())
```

```python
#getting dummy variables for 'Reason for Absence' column for every unique data in this column
reason_data = pd.get_dummies(df["Reason for Absence"]).astype(int)
reason_data.head()
```

```python
#create another column ,to ensure there is only one true value in each row.
reason_data['check'] = reason_data.sum(axis = 1)
reason_data.head()
```

```python
#checking if the total sum of the "check" column is equal to the number of total rows
reason_data['check'].sum()
```

```python
reason_data = reason_data.drop(['check'],axis = 1)
reason_data.head()
```

```python
#recreate the reason_data column
reason_data = pd.get_dummies(df["Reason for Absence"],drop_first=True).astype(int)
reason_data.head()
```

```python
reason_data.columns.values
```

```python
reason_type_1 = reason_data.loc[:,1:14].max(axis = 1)
reason_type_2 = reason_data.loc[:,15:17].max(axis = 1)
reason_type_3 = reason_data.loc[:,18:21].max(axis = 1)
reason_type_4 = reason_data.loc[:,22:].max(axis = 1)
```

```python
reason_type_1.head(10)
```

```python
#adding 4 new reason type with df(main dataset)
df = pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis = 1)
df.head()
```

```python
df = df.drop(['Reason for Absence'],axis =1)
df.head()
```

```python
df.columns.values
```

```python
columns_name = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_type_1','Reason_type_2','Reason_type_3','Reason_type_4']
```

## Creating checkpoints

```python
df_mod = df.copy()
df_mod.head()
```

## Working with Date column

```python
type(df_mod['Date'][0])
```

```python
df_mod['Date'] = pd.to_datetime(df_mod['Date'],format = '%d/%m/%Y')
df_mod.info()
```

### extracting month value from the data


```python
list_month = []
for i in range(df_mod.shape[0]):
    list_month.append(df_mod['Date'][i].month)
```

```python
df_mod['Month'] = list_month
sorted(df_mod['Month'].unique())
```

## extracting the weekday from date

```python
#creating a function for retrieving weekday
```

```python
df_mod['Date'][0].weekday()
```

```python
def date_to_weekday(Date):
    return Date.weekday()
```

```python
#adding weekday column into df_mod with the weekday value for each week
df_mod['Weekday'] = df_mod['Date'].apply(date_to_weekday)
df_mod.head(10)
```

```python
#dropping the date column
df_mod = df_mod.drop(['Date'],axis=1)
df_mod.head()
```

## Creating Checkpoints 

```python
df_mod_2 = df_mod.copy()
df_mod_2.head()
```

## Working with Education column
    Turn the data from the ‘Education’ column into binary data, by mapping the value of 0 to the value of 1, and the value of 1 to 
the rest of the values found in this column.

```python
df_mod_2['Education'].unique()
```

```python
df_mod_2['Education'].value_counts()
```

```python
df_mod_2['Education'] = df_mod['Education'].map({1:0,2:1,3:1,4:1})
df_mod['Education'].value_counts()
```
